```python
# 调用服务接口添加 question 的示例

import requests
import json

# 服务地址（根据实际情况修改）
BASE_URL = "http://127.0.0.1:80"  # 本地调试地址
BASE_URL = "https://api.weixin.qq.com"
# BASE_URL = "https://your-domain.com"  # 生产环境地址

# 添加问题的接口地址
API_URL = f"{BASE_URL}/api/questions"
```

```python
response = requests.get(
    f"{BASE_URL}/api/companies",
    headers={"Content-Type": "application/json"}
)

print(f"状态码: {response.status_code}")
# print(f"响应内容: {response.json()}")
```

```python
response.json()
```

```python
for item in response.json()["data"]:
    print(item["name"])
```

```python
response = requests.get(
    BASE_URL+"/api/question_detail"+"?question_id=528",
    headers={"Content-Type": "application/json"}
)

print(f"状态码: {response.status_code}")
print(f"响应内容: {response.json()}")
```
